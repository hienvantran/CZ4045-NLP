## Import dependancies

```python
import vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/ryanphuay/Desktop/CZ4045-NLP/Crawled_data/8k-unlabeled.csv')
df.head(20)
```

```python
analyzer = SentimentIntensityAnalyzer()


df['neg'] = [analyzer.polarity_scores(x)['neg'] for x in df['title']]
df['neu'] = [analyzer.polarity_scores(x)['neu'] for x in df['title']]
df['pos'] = [analyzer.polarity_scores(x)['pos'] for x in df['title']]

df['compound'] = [analyzer.polarity_scores(x)['compound'] for x in df['title']] # normalised compounded valence score

df.head(20)
```

```python
df.loc[df.compound == 0, 'sentiment'] = 'neutral'
df.loc[df.compound > 0, 'sentiment'] = 'positive'
df.loc[df.compound < 0, 'sentiment'] = 'negative'
df.head()
```

```python
df.to_csv('/Users/ryanphuay/Desktop/CZ4045-NLP/Crawled_data/8k-labelled-vader.csv', index=False)
```
